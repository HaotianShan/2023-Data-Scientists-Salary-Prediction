# Data Cleaning

```python
df.isnull().sum()/df.shape[0]*100
```

#### Replacing into full form of experience_level and employment_type columns

```python
df['experience_level']=df['experience_level'].str.replace('SE','Senior-level').str.replace('MI','Mid-level').str.replace('EN','Entry-level').str.replace('EX','Executive-level')
df['employment_type']=df['employment_type'].str.replace('PT','Part-time').str.replace('FT','Full-time').str.replace('CT','Contract').str.replace('FL','Freelance')
```

#### Replacing less occur values of company_location into other

```python
replace_specific = {
    'US':'US',
    'GB':'GB',
    "CA":'CA',
    'ES':'ES',
    'IN':'IN',
    'DE':'DE',
    'FR':'FR'
}
other_value = 'Other'
df['company_location'] = df['company_location'].where(df['company_location'].isin(replace_specific.values()), other_value)

```

#### Dropping some features

```python
df.drop(['salary_currency','salary','remote_ratio','employee_residence'],axis=1,inplace=True)
```

#### Only taking Data Engineer, Data Scientist, Data Analyst,Machine Learning Engineer and Analytics Engineer job title

```python
mask1=df['job_title']=="Data Engineer"
mask2=df['job_title']=="Data Scientist"
mask3=df['job_title']=="Data Analyst"
mask4=df['job_title']=="Machine Learning Engineer"
mask5=df['job_title']=="Analytics Engineer"
mask6=df['job_title']=="Data Architect"

#df = df.loc[mask1 | mask2 | mask3 | mask4 | mask5]

df=df.loc[mask1 | mask2 | mask3 | mask4 | mask5 | mask6 ,['work_year', 'experience_level', 'employment_type', 'job_title',
       'salary_in_usd','company_location', 'company_size']]
df.shape
df.reset_index(drop='True',inplace=True)
```

# Encoding

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder

allCatCol=list(df.select_dtypes(include="object").columns)

categorical_column=None

categorical_column=['experience_level', 'employment_type', 'job_title', 'company_location', 'company_size']




df["company_size"]=df["company_size"].map({'M': 0, 'L': 1, 'S': 2})

encoding= OneHotEncoder(sparse=False)
#*********** Performing one-Hot encoding ************
result_encod=encoding.fit_transform(df[['experience_level', 'employment_type', 'job_title', 'company_location']])
dataFrame_encode_col=pd.DataFrame(result_encod,columns=['experience_level_Entry-level', 'experience_level_Executive-level', 'experience_level_Mid-level', 'experience_level_Senior-level', 'employment_type_Contract', 'employment_type_Freelance', 'employment_type_Full-time', 'employment_type_Part-time', 'job_title_Analytics Engineer', 'job_title_Data Analyst', 'job_title_Data Architect', 'job_title_Data Engineer', 'job_title_Data Scientist', 'job_title_Machine Learning Engineer', 'company_location_CA', 'company_location_DE', 'company_location_ES', 'company_location_FR', 'company_location_GB', 'company_location_IN', 'company_location_Other', 'company_location_US'])
numeric_col=df[['work_year', 'salary_in_usd', 'company_size']]
final_dataset=pd.merge(dataFrame_encode_col,numeric_col,left_index=True,right_index=True)
```

```python
final_dataset.head()
```

# Model Training 

```python
x=final_dataset.drop('salary_in_usd', axis=1)
y=final_dataset['salary_in_usd']
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1, random_state=21)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor as randomForestRegressor

rfr=randomForestRegressor() 
rfr.fit(x_train,y_train)
rfrtd=rfr.predict(x_test)
rfrr2=r2_score(y_test,rfrtd)
print('r2-score',rfrr2)

y_pred = rfr.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)

rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)

mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error (MAE):", mae)
```

#### Plot the residuals

```python
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.show()
```
